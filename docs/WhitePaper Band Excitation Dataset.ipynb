## Band Excitation Dataset

Band Excitation is a technique used in atomic force microscopy (AFM) that involves applying a sinusoidal excitation signal to the cantilever probe of the microscope over a range of frequencies.[1] This allows for the simultaneous collection of information at multiple frequencies, enabling a more complete analysis of material properties and surface interactions. However, despite its potential, there is a significant challenge in processing the voluminous data generated by Band Excitation experiments. This has hindered the translation of a substantial portion of the acquired information into usable knowledge, creating the need for advanced computational methodologies to maximize this technique.

### Description

 ##### Structure
 The data is often structured as a series of frequency-dependent force-distance curves or spectra. Each curve corresponds to a specific frequency of excitation.
 
 ##### Format
 Band excitation data is commonly stored in a digital format, often as multi-dimensional arrays or matrices. 
 

### Source
This dataset was obtained from the Multifunctional Materials and Machine Learning Group(M3). 

## Problem Statement

Band excitation data frequently yields a high volume of information; however, conventional scientific methods encounter challenges in effectively translating this data into actionable insights. By leveraging deep learning models in machine learning, we can discern critical features influencing ferroelectric switching. This not only establishes a foundation for understanding and predicting domain structure transformations in ferroelectric materials but also lays the groundwork for their manipulation and validation.

## State-of-the-Art Review

In the domain of controlling and verifying domain structure transformations using Band Excitation data, state-of-the-art methodologies primarily revolve around the integration of advanced machine learning techniques. Researchers are employing deep learning architectures, such as convolutional and recurrent neural networks, for accurate prediction and control of domain structure transformations. These models are capable of processing complex Band Excitation data, enabling precise manipulation and validation of domain structures.

The paper "Revealing ferroelectric switching character using deep recurrent neural networks" by Joshua C. Agar, it delves into the application of deep recurrent neural networks for understanding ferroelectric switching behavior. The study employs advanced neural network architectures to extract intricate patterns in the data, shedding light on the underlying mechanisms of ferroelectric materials and opening avenues for enhanced control in domain structure transformations.

This whitepaper demonstrates the potential of integrating Machine Learning with Band Excitation data to delve into ferroelectric material behavior at the nanoscale. The insights gained have far-reaching implications for materials science, technology, and engineering applications.
